Functions in python

```python
# Hello world function
def hello():  #defining the function
    print('Hello Everyone')
hello()  # calling the function . We can call a function multiple times

# add function
def add(x,y):
    c = x+y
    return c   #return statement

x = add(5,4)
print(x)

```

```python
#square root function by taking input from user
def squart(x):
    root = x**(1/2)
    return root

a = int(input("enter the number "))
print(squart(a))
```

```python
# returning two variables
def add_sub(x,y):
    c = x+y
    d = x-y
    return c,d
#taking multiple integers as input using map function
a,b = map(int, input("enter the numbers separated by comma ").split(','))
result = add_sub(a,b)
print(result)
    
```

There are 6 types of plots
1. Bar graph

```python
from matplotlib import pyplot as plt
from matplotlib import style

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]

plt.bar(x1,y1,color='g',label = 'bar one')
plt.bar(x2,y2,color='r',label = 'bar two')

plt.title('sample')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend(loc= 'lower right')

plt.grid(True,color='k')
plt.show()
```

2. Histogram

```python
import matplotlib.pyplot as plt

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]

plt.hist(x1, y1,label = 'first one', histtype = 'bar', rwidth = 0.8)
plt.hist(x2, y2,label = 'second one', histtype = 'bar', rwidth = 0.8)
plt.xlabel('X')
plt.ylabel('Y')

plt.title('Histogram')

plt.legend(loc = 'upper right')

plt.grid(True,color='k')

plt.show()
```

3. Scatter plot

```python
import matplotlib.pyplot as plt

x1 = [2,3,4,5,6,7,8,20,50]          
y1 = [4,8,10,30,60,70,80,90,100]

x2 = [5,10,39]
y2 = [6,10,20]


plt.scatter(x1,y1,label= 'skt1', color = 'b',s=25,marker = 'o')
plt.scatter(x2,y2,label= 'skt2', color = 'r',s=25,marker = 'o')
plt.xlabel('X')
plt.ylabel('Y')

plt.title('scatter plot')
plt.legend()
plt.show()
```

4. Stack Plot

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]     #here in this the variables ie the list must be of same length.     
y1 = [2,3,4,3,2]
x2 = [5,10,39,4,3]   
y2 = [6,10,20,4,5]

plt.plot([],[],color='m',label='x2', linewidth = 5)
plt.plot([],[],color='c',label='y1', linewidth = 5)
plt.plot([],[],color='r',label='y2', linewidth = 5)

plt.stackplot(x1,x2,y1,y2,colors = ['m','c','r'])

plt.xlabel('X')
plt.ylabel('Y')

plt.title('Area plot')
plt.legend()
plt.grid(True,color = 'k')
plt.show()



```

5. Pie Charts

```python
import matplotlib.pyplot as plt

x = [3,4,2,6]

plt.pie(x,labels=['sleeping','eating','playing','working'],colors= ['m','r','b','g'],startangle = 45,shadow=True,explode=(0,0,0,0),autopct='%.2f')

plt.title('pie chart')
plt.show()
```

Working with multiple Plots

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.subplot(221)     #221 means - 2(graphs horizontally), 2(total 2 graphs), 1(this is the first graph)
plt.plot(t1, f(t1), 'bo', t2, f(t2))

plt.subplot(222)
plt.plot(t2, f(t2), np.cos(2*np.pi*t2))
plt.show()
```
